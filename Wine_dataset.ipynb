```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import plotly.express as px
df=pd.read_csv('/content/winequalityN.csv')
df.head(50)
```

```python
x=df.density
y=df.pH
plt.plot(y,color="yellow")
plt.title('LINE PLOT')
```

```python
z=df.quality
plt.scatter(x,z,color="purple")
plt.title('SCATTER PLOT')
```

```python
df1=df.head(10)
a=df1['fixed acidity']
label=df1['sulphates']
plt.pie(a,labels=label)
plt.title('PIE CHART')
```

```python
plt.boxplot(a)
plt.title('BOX PLOT')
```

```python
plt.stem(z,x)
plt.title('LOLLIPOP CHART')
```

```python
plt.stackplot(df1['fixed acidity'],df1['chlorides'],color="red")
plt.title('STACKPLOT')
```

```python
plt.bar(y,z,color="black")
plt.title('BAR CHART')
```
